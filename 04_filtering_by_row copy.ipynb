```python
import numpy as np
import pandas as pd

import env
```

```python
url = env.get_db_url('employees')
sql = 'SELECT * FROM employees'

df = pd.read_sql(sql, url)
```

```python
df.head()
```

Find all current or previous employees with first names 'Irena', 'Vidya', or 'Maya'

```python
df[df.first_name.isin(['Irena', 'Vidya', 'Maya'])]
```

Find all current or previous employees with first names 'Irena', 'Vidya', or 'Maya', and who is male.

```python
df[(df.first_name.isin(['Irena', 'Vidya', 'Maya'])) 
  & (df.gender == 'M')]
```

Find all current or previous employees whose last name starts with 'E'

```python
df[df.last_name.str.startswith('E')]
```

Find all current or previous employees whose last name starts with 'E' or ends with 'e'.

```python
df[
    (df.last_name.str.startswith('E'))
   |(df.last_name.str.endswith('e'))
  ]
```

Find all current or previous employees whose last name starts with 'E' and ends with 'e'.

```python
df[
    (df.last_name.str.startswith('E'))
   &(df.last_name.str.endswith('e'))
  ]
```

-- Find all current or previous employees hired in the 90s.


```python
df[df.hire_date.astype(str).str.startswith('199')]
```

-- Find all current or previous employees born on Christmas.


```python
df[df.birth_date.astype(str).str.endswith('-12-25')]
```

-- Find all current or previous employees hired in the 90s and born on Christmas.

```python
df[
    (df.birth_date.astype(str).str.endswith('-12-25'))
   &(df.hire_date.astype(str).str.startswith('199')) 
]
```

-- Find all current or previous employees with a 'q' in their last name. 


```python
df[df.last_name.str.lower().str.contains('q')]
```

-- Find all current or previous employees with a 'q' in their last name but not 'qu'


```python
df[
    (df.last_name.str.lower().str.contains('q'))
  & (~df.last_name.str.lower().str.contains('qu'))
    ]
```
